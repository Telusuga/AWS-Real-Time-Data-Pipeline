```python
c=Consumer(login_info())
```

```python
topic='aws'
c.subscribe([topic])
```

```python
s3 = boto3.client('s3')
```

```python
bucket_name = 'kafkaconsumer01'
```

```python
while True:
    try:
        msg = c.poll(5.0)

        if msg is None:
            print('Waiting for messages!!')
            continue
        elif msg is not None:
            msg_value = msg.value().decode('utf-8')
            obj_key=str(uuid.uuid4())+'.json'
            print(msg_value)
            print(obj_key)
            response = s3.put_object(
            Body=msg_value,
            Bucket=bucket_name,
            Key=obj_key
            )
            if response['ResponseMetadata']['HTTPStatusCode'] == 200:
                print('String value inserted into S3 bucket successfully.')
            else:
                print('Error inserting the string value into S3 bucket.')

    except KeyboardInterrupt:
        print('Closing the consumer code')
        break



```
